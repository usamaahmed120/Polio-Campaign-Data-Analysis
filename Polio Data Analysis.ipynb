# Form2A Data Validation

```python
# importing libraries thar are required
import pandas as pd
import os
writer = pd.ExcelWriter('Output/form2a_validation_matching.xlsx', engine='xlsxwriter')
```

### Conditions/Functions Starts Here

```python
# For House Hold Microplan
def hh_microplan(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 130 or x <= 5:
        return color
    else:
        return default

# For House Hold Visited
def hh_visited(x):
    color = 'background-color: yellow'
    default = ''
    if x['Number of HH in Microplan'] - x['Number of HH Visited'] >= 30 or x['Number of HH in Microplan'] - x['Number of HH Visited'] <= -30:
        return [default,color]
    else:
        return [default,default]
    
# For Daily Target Children 0-11
def daily_target_children_1(x):
    color = 'background-color: yellow'
    default = ''
    if (x['Daily Target Children 0-11 M'] - (x['Vaccinated (A1) 0-11 M'] + x['Recorded Missed NA/L 0-11 M'] + x['Recorded Missed Refusal 0-11 M']) >= 20) or (x['Daily Target Children 0-11 M'] - (x['Vaccinated (A1) 0-11 M'] + x['Recorded Missed NA/L 0-11 M'] + x['Recorded Missed Refusal 0-11 M']) <= -20):
        return [color,default,default,default]
    else:
        return [default,default,default,default]
    
# For Daily Target Children 12-59
def daily_target_children_2(x):
    color = 'background-color: yellow'
    default = ''
    if (x['Daily Target Children 12-59 M'] - (x['Vaccinated (A1) 12-59 M'] + x['Recorded Missed NA/L 12-59 M'] + x['Recorded Missed Refusal 12-59 M']) >= 20) or (x['Daily Target Children 12-59 M'] - (x['Vaccinated (A1) 12-59 M'] + x['Recorded Missed NA/L 12-59 M'] + x['Recorded Missed Refusal 12-59 M']) <= -20):
        return [color,default,default,default]
    else:
        return [default,default,default,default]
    
# For Children Recorded at HH today 0-11 M
def hh_today_1(x):
    color = 'background-color: yellow'
    default = ''
    if x!=0:
        return color
    else:
        return default
    
# For Children Recorded at HH today 12-59 M
def hh_today_2(x):
    color = 'background-color: yellow'
    default = ''
    if x!=0:
        return color
    else:
        return default
    
# For Recorded Missed NA/L 0-11 M
def r1_na(x):
    color = 'background-color: yellow'
    default = ''
    if x['Recorded Missed NA/L 0-11 M'] < x['Covered Same Day NA/L 0-11 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Recorded Missed NA/L 12-59 M
def r2_na(x):
    color = 'background-color: yellow'
    default = ''
    if x['Recorded Missed NA/L 12-59 M'] < x['Covered Same Day NA/L 12-59 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Recorded Missed Refusal 0-11 M
def r1_refusal(x):
    color = 'background-color: yellow'
    default = ''
    if x['Recorded Missed Refusal 0-11 M'] < x['Covered Same Day Refusal 0-11 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Recorded Missed Refusal 12-59 M
def r2_refusal(x):
    color = 'background-color: yellow'
    default = ''
    if x['Recorded Missed Refusal 12-59 M'] < x['Covered Same Day Refusal 12-59 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Covered Same Day NA/L 0-11 M
def c1_na(x):
    color = 'background-color: yellow'
    default = ''
    if x['Covered Same Day NA/L 0-11 M'] > x['Recorded Missed NA/L 0-11 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Covered Same Day NA/L 12-59 M
def c2_na(x):
    color = 'background-color: yellow'
    default = ''
    if x['Covered Same Day NA/L 12-59 M'] > x['Recorded Missed NA/L 12-59 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Covered Same Day Refusal 0-11 M
def c1_refusal(x):
    color = 'background-color: yellow'
    default = ''
    if x['Covered Same Day Refusal 0-11 M'] > x['Recorded Missed Refusal 0-11 M']:
        return [color, default]
    else:
        return[default,default]
    
# For Covered Same Day Refusal 12-59 M
def c2_refusal(x):
    color = 'background-color: yellow'
    default = ''
    if x['Covered Same Day Refusal 12-59 M'] > x['Recorded Missed Refusal 12-59 M']:
        return [color, default]
    else:
        return[default,default]
    
# For School/Madrassa Children

def school_madrassa(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 130:
        return color
    else:
        return default
    
# For Guest (G) vaccinated
def guest_children(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 30:
        return color
    else:
        return default
    
# For Children Vaccinated on Streets
def street_children(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 30:
        return color
    else:
        return default
    
# For HRMP (High Risk Mobile Population)
def hrmp(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 130:
        return color
    else:
        return default
    
# For AFP
def accute(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 3:
        return color
    else:
        return default
    
# For Zero Dose Total
def zdose(x):
    color = 'background-color: yellow'
    default = ''
    if x >= 20:
        return color
    else:
        return default
    
# For Doses Received
def dose_rec(x):
    color = 'background-color: yellow'
    default = ''
    if x['Doses Received'] < x['Doses Used'] or x['Doses Received'] < x['Doses Returned'] :
        return [color, default, default]
    else:
        return [default, default, default]
    
# For Doses Used
def dose_used(x):
    color = 'background-color: yellow'
    default = ''
    if x['Doses Used'] > x['Doses Received']:
        return [color, default]
    else:
        return [default, default]
    
# For Doses Returned
def dose_returned(x):
    color = 'background-color: yellow'
    default = ''
    if (x['Doses Received'] - x['Doses Used']) != x['Doses Returned']:
        return [color, default, default]
    else:
        return [default, default, default]
    
# For Comparision Between Doses Used and Total No Of Vaccinated Child
def dose_vchild(x):
    color = 'background-color: pink'
    default = ''
    if ((x['Vaccinated (A1) 0-11 M'] + x['Vaccinated (A1) 12-59 M'] + x['Covered Same Day NA/L 0-11 M'] + x['Covered Same Day NA/L 12-59 M'] + x['Covered Same Day Refusal 0-11 M'] + x['Covered Same Day Refusal 12-59 M'] + x['School/Madrassa Children'] + x['Guest (G) vaccinated'] + x['Children Vaccinated in Streets']) - x['Doses Used']) >=30 or ((x['Vaccinated (A1) 0-11 M'] + x['Vaccinated (A1) 12-59 M'] + x['Covered Same Day NA/L 0-11 M'] + x['Covered Same Day NA/L 12-59 M'] + x['Covered Same Day Refusal 0-11 M'] + x['Covered Same Day Refusal 12-59 M'] + x['School/Madrassa Children'] + x['Guest (G) vaccinated']) - x['Doses Used']) <=-30 :
        return [default, default, default, default, default, default, default, default, default, color]
    else:
        return [default, default, default, default,default,default,default,default,default,default]
```

```python
# Making function for matching

def na_matching(x):
    default = ''
    color = 'background-color: yellow'
    if x['Form2A SMC NA/L'] != x['Not Available']:
        return [color,color]
    else:
        return [default,default]

def r_matching(x):
    default = ''
    color = 'background-color: yellow'
    if x['Form2A Refusal'] != x['Refused']:
        return [color,color]
    else:
        return [default,default]

# Making function for ZeroDose

def zerodose_matching(x):
    default = ''
    color = 'background-color: yellow'
    if x['Form2A Zerodose'] != x['Zerodose']:
        return [color,color]
    else:
        return [default,default]  

# Making function for AFP   

def afp_matching(x):
    default = ''
    color = 'background-color: yellow'
    if x['Form2A AFP'] != x['AFP']:
        return [color,color]
    else:
        return [default,default]
```

### Conditions/Functions Starts To Apply On Data Here

```python
if os.path.exists('Data/form2a.csv') and os.path.exists('Data/linelist.csv') and os.path.exists('Data/zerodose.csv') and os.path.exists('Data/afp.csv'):   
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    linelist = pd.read_csv('Data/linelist.csv',dtype={'Coverage date/time':'str'})
    zerodose = pd.read_csv('Data/zerodose.csv')
    afp = pd.read_csv('Data/afp.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    # Linelist

    # to fill nan value with empty
    linelist = linelist.fillna('')
    linelist['Reason Missed'] = linelist['Reason Missed'].apply(lambda x: 'Not Available' if x == 'Locked House' else x )
    # concatenating values/making new column for unique key
    linelist['Unique Key'] = linelist['Campaign'] + linelist['Province'] + linelist['District'] + linelist['Tehsil'] + linelist['UC'] + linelist['TID'] + linelist['Area Incharge'] + linelist['Day'].astype(str) + linelist['Team No'].astype(str) + linelist['Team Member']
    # transforming data
    linelist = linelist.pivot_table(index ='Unique Key', columns='Reason Missed', values='Campaign', aggfunc='count',fill_value=0)
    # reset index
    linelist.reset_index(inplace=True)
    # Joining Form2A Table with Linelist
    linelist = form2a.merge(linelist,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']],right_on='Unique Key')
    # Calculating SMC in Form2A
    linelist['Form2A SMC NA/L'] = (linelist['Recorded Missed NA/L 0-11 M'] + linelist['Recorded Missed NA/L 12-59 M']) - (linelist['Covered Same Day NA/L 0-11 M'] + linelist['Covered Same Day NA/L 12-59 M'])
    linelist['Form2A Refusal'] = (linelist['Recorded Missed Refusal 0-11 M'] + linelist['Recorded Missed Refusal 12-59 M']) - (linelist['Covered Same Day Refusal 0-11 M'] + linelist['Covered Same Day Refusal 12-59 M'])
    #fill nan values with 0
    linelist.fillna(0,inplace=True)
    # Arranging Column
    linelist = linelist.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Form2A SMC NA/L','Form2A Refusal','Zero Dose Total','AFP','Not Available','Refused'])
    # Renaming columns
    linelist.rename(columns={'Zero Dose Total':'Form2A Zerodose','AFP':'Form2A AFP'}, inplace=True)
    
 
    
    # ZeroDose
    
    # Filling NaN values with empty
    zerodose.fillna('',inplace=True)
    # Concatenating values
    zerodose['Unique Key'] = zerodose['Campaign'] + zerodose['Province'] + zerodose['District'] + zerodose['Tehsil'] + zerodose['UC'] + zerodose['TID'] + zerodose['Area Incharge'] + zerodose['Day'].astype(str) + zerodose['Team No'].astype(str) + zerodose['Team Member']
    # renaming campaign name to zerodose
    zerodose.rename(columns={'Campaign':'Zerodose'},inplace=True)
    # Transforming data
    zerodose = zerodose.pivot_table(index='Unique Key', values='Zerodose', aggfunc='count',fill_value=0)
    # Reseting index
    zerodose.reset_index(inplace=True)
    # Merging Data with linelist
    zerodose = linelist.merge(zerodose,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    zerodose.fillna(0,inplace=True)
    # Rearraning columns
    linelist['Zerodose'] = zerodose.reindex(columns=['Zerodose'])
    
    
    
    # AFP
    
    # Filling NaN values with empty
    afp.fillna('',inplace=True)
    # Concatenating values
    afp['Unique Key'] = afp['Campaign'] + afp['Province'] + afp['District'] + afp['Tehsil'] + afp['UC'] + afp['TID'] + afp['Area Incharge'] + afp['Day'].astype(str) + afp['Team No'].astype(str) + afp['Team Member']
    # renaming campaign name to zerodose
    afp.rename(columns={'Campaign':'AFP'},inplace=True)
    # Transforming data
    afp = afp.pivot_table(index='Unique Key', values='AFP', aggfunc='count',fill_value=0)
    # Reseting index
    afp.reset_index(inplace=True)
    # Merging Data with linelist
    afp = linelist.merge(afp,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    afp.fillna(0,inplace=True)
    # Rearraning columns
    linelist['AFP'] = afp.reindex(columns=['AFP'])
    
    
    # Applying all function here
    linelist.style.\
    apply(na_matching,axis=1,subset=['Form2A SMC NA/L','Not Available']).\
    apply(r_matching,axis=1,subset=['Form2A Refusal','Refused']).\
    apply(zerodose_matching,axis=1,subset=['Form2A Zerodose','Zerodose']).\
    apply(afp_matching,axis=1,subset=['Form2A AFP','AFP']).to_excel(writer,sheet_name='matching',index=False)
    

elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/linelist.csv') and os.path.exists('Data/zerodose.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    linelist = pd.read_csv('Data/linelist.csv',dtype={'Coverage date/time':'str'})
    zerodose = pd.read_csv('Data/zerodose.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    # Linelist

    # to fill nan value with empty
    linelist = linelist.fillna('')
    linelist['Reason Missed'] = linelist['Reason Missed'].apply(lambda x: 'Not Available' if x == 'Locked House' else x )
    # concatenating values/making new column for unique key
    linelist['Unique Key'] = linelist['Campaign'] + linelist['Province'] + linelist['District'] + linelist['Tehsil'] + linelist['UC'] + linelist['TID'] + linelist['Area Incharge'] + linelist['Day'].astype(str) + linelist['Team No'].astype(str) + linelist['Team Member']
    # transforming data
    linelist = linelist.pivot_table(index ='Unique Key', columns='Reason Missed', values='Campaign', aggfunc='count',fill_value=0)
    # reset index
    linelist.reset_index(inplace=True)
    # Joining Form2A Table with Linelist
    linelist = form2a.merge(linelist,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']],right_on='Unique Key')
    # Calculating SMC in Form2A
    linelist['Form2A SMC NA/L'] = (linelist['Recorded Missed NA/L 0-11 M'] + linelist['Recorded Missed NA/L 12-59 M']) - (linelist['Covered Same Day NA/L 0-11 M'] + linelist['Covered Same Day NA/L 12-59 M'])
    linelist['Form2A Refusal'] = (linelist['Recorded Missed Refusal 0-11 M'] + linelist['Recorded Missed Refusal 12-59 M']) - (linelist['Covered Same Day Refusal 0-11 M'] + linelist['Covered Same Day Refusal 12-59 M'])
    #fill nan values with 0
    linelist.fillna(0,inplace=True)
    # Arranging Column
    linelist = linelist.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Form2A SMC NA/L','Form2A Refusal','Zero Dose Total','Not Available','Refused'])
    # Renaming columns
    linelist.rename(columns={'Zero Dose Total':'Form2A Zerodose'}, inplace=True)
    
 
    
    # ZeroDose
    
    # Filling NaN values with empty
    zerodose.fillna('',inplace=True)
    # Concatenating values
    zerodose['Unique Key'] = zerodose['Campaign'] + zerodose['Province'] + zerodose['District'] + zerodose['Tehsil'] + zerodose['UC'] + zerodose['TID'] + zerodose['Area Incharge'] + zerodose['Day'].astype(str) + zerodose['Team No'].astype(str) + zerodose['Team Member']
    # renaming campaign name to zerodose
    zerodose.rename(columns={'Campaign':'Zerodose'},inplace=True)
    # Transforming data
    zerodose = zerodose.pivot_table(index='Unique Key', values='Zerodose', aggfunc='count',fill_value=0)
    # Reseting index
    zerodose.reset_index(inplace=True)
    # Merging Data with linelist
    zerodose = linelist.merge(zerodose,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    zerodose.fillna(0,inplace=True)
    # Rearraning columns
    linelist['Zerodose'] = zerodose.reindex(columns=['Zerodose'])
    
    
    # Applying all function here
    linelist.style.\
    apply(na_matching,axis=1,subset=['Form2A SMC NA/L','Not Available']).\
    apply(r_matching,axis=1,subset=['Form2A Refusal','Refused']).\
    apply(zerodose_matching,axis=1,subset=['Form2A Zerodose','Zerodose']).to_excel(writer,sheet_name='matching',index=False)


    
elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/linelist.csv') and os.path.exists('Data/afp.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    linelist = pd.read_csv('Data/linelist.csv',dtype={'Coverage date/time':'str'})
    afp = pd.read_csv('Data/afp.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    # Linelist

    # to fill nan value with empty
    linelist = linelist.fillna('')
    linelist['Reason Missed'] = linelist['Reason Missed'].apply(lambda x: 'Not Available' if x == 'Locked House' else x )
    # concatenating values/making new column for unique key
    linelist['Unique Key'] = linelist['Campaign'] + linelist['Province'] + linelist['District'] + linelist['Tehsil'] + linelist['UC'] + linelist['TID'] + linelist['Area Incharge'] + linelist['Day'].astype(str) + linelist['Team No'].astype(str) + linelist['Team Member']
    # transforming data
    linelist = linelist.pivot_table(index ='Unique Key', columns='Reason Missed', values='Campaign', aggfunc='count',fill_value=0)
    # reset index
    linelist.reset_index(inplace=True)
    # Joining Form2A Table with Linelist
    linelist = form2a.merge(linelist,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']],right_on='Unique Key')
    # Calculating SMC in Form2A
    linelist['Form2A SMC NA/L'] = (linelist['Recorded Missed NA/L 0-11 M'] + linelist['Recorded Missed NA/L 12-59 M']) - (linelist['Covered Same Day NA/L 0-11 M'] + linelist['Covered Same Day NA/L 12-59 M'])
    linelist['Form2A Refusal'] = (linelist['Recorded Missed Refusal 0-11 M'] + linelist['Recorded Missed Refusal 12-59 M']) - (linelist['Covered Same Day Refusal 0-11 M'] + linelist['Covered Same Day Refusal 12-59 M'])
    #fill nan values with 0
    linelist.fillna(0,inplace=True)
    # Arranging Column
    linelist = linelist.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Form2A SMC NA/L','Form2A Refusal','AFP','Not Available','Refused'])
    # Renaming columns
    linelist.rename(columns={'AFP':'Form2A AFP'}, inplace=True)
    
    
    
    # AFP
    
    # Filling NaN values with empty
    afp.fillna('',inplace=True)
    # Concatenating values
    afp['Unique Key'] = afp['Campaign'] + afp['Province'] + afp['District'] + afp['Tehsil'] + afp['UC'] + afp['TID'] + afp['Area Incharge'] + afp['Day'].astype(str) + afp['Team No'].astype(str) + afp['Team Member']
    # renaming campaign name to zerodose
    afp.rename(columns={'Campaign':'AFP'},inplace=True)
    # Transforming data
    afp = afp.pivot_table(index='Unique Key', values='AFP', aggfunc='count',fill_value=0)
    # Reseting index
    afp.reset_index(inplace=True)
    # Merging Data with linelist
    afp = linelist.merge(afp,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    afp.fillna(0,inplace=True)
    # Rearraning columns
    linelist['AFP'] = afp.reindex(columns=['AFP'])
    
    
    # Applying all function here
    linelist.style.\
    apply(na_matching,axis=1,subset=['Form2A SMC NA/L','Not Available']).\
    apply(r_matching,axis=1,subset=['Form2A Refusal','Refused']).\
    apply(afp_matching,axis=1,subset=['Form2A AFP','AFP']).to_excel(writer,sheet_name='matching',index=False)

elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/zerodose.csv') and os.path.exists('Data/afp.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    zerodose = pd.read_csv('Data/zerodose.csv')
    afp = pd.read_csv('Data/afp.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    # ZeroDose
    
    # Filling NaN values with empty
    zerodose.fillna('',inplace=True)
    # Concatenating values
    zerodose['Unique Key'] = zerodose['Campaign'] + zerodose['Province'] + zerodose['District'] + zerodose['Tehsil'] + zerodose['UC'] + zerodose['TID'] + zerodose['Area Incharge'] + zerodose['Day'].astype(str) + zerodose['Team No'].astype(str) + zerodose['Team Member']
    # renaming campaign name to zerodose
    zerodose.rename(columns={'Campaign':'Zerodose'},inplace=True)
    # Transforming data
    zerodose = zerodose.pivot_table(index='Unique Key', values='Zerodose', aggfunc='count',fill_value=0)
    # Reseting index
    zerodose.reset_index(inplace=True)
    # Merging Data with linelist
    zerodose = form2a.merge(zerodose,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    zerodose.fillna(0,inplace=True)
    # Rearraning columns
    zerodose = zerodose.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Zero Dose Total','AFP','Zerodose'])

    
    
    # AFP
    
    # Filling NaN values with empty
    afp.fillna('',inplace=True)
    # Concatenating values
    afp['Unique Key'] = afp['Campaign'] + afp['Province'] + afp['District'] + afp['Tehsil'] + afp['UC'] + afp['TID'] + afp['Area Incharge'] + afp['Day'].astype(str) + afp['Team No'].astype(str) + afp['Team Member']
    # renaming campaign name to zerodose
    afp.rename(columns={'Campaign':'Linelist AFP'},inplace=True)
    # Transforming data
    afp = afp.pivot_table(index='Unique Key', values='Linelist AFP', aggfunc='count',fill_value=0)
    # Reseting index
    afp.reset_index(inplace=True)
    # Merging Data with linelist
    afp = form2a.merge(afp,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    afp.fillna(0,inplace=True)
    # Rearraning columns
    zerodose['Linelist AFP'] = afp.reindex(columns=['Linelist AFP'])
    # Renaming column names
    zerodose.rename(columns={'Zero Dose Total':'Form2A Zerodose','AFP':'Form2A AFP', 'Linelist AFP':'AFP'}, inplace=True)
    
    # Applying all function here
    zerodose.style.\
    apply(afp_matching,axis=1,subset=['Form2A AFP','AFP']).\
    apply(zerodose_matching,axis=1,subset=['Form2A Zerodose','Zerodose']).to_excel(writer,sheet_name='matching',index=False)


elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/linelist.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    linelist = pd.read_csv('Data/linelist.csv',dtype={'Coverage date/time':'str'})
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    # Linelist

    # to fill nan value with empty
    linelist = linelist.fillna('')
    linelist['Reason Missed'] = linelist['Reason Missed'].apply(lambda x: 'Not Available' if x == 'Locked House' else x )
    # concatenating values/making new column for unique key
    linelist['Unique Key'] = linelist['Campaign'] + linelist['Province'] + linelist['District'] + linelist['Tehsil'] + linelist['UC'] + linelist['TID'] + linelist['Area Incharge'] + linelist['Day'].astype(str) + linelist['Team No'].astype(str) + linelist['Team Member']
    # transforming data
    linelist = linelist.pivot_table(index ='Unique Key', columns='Reason Missed', values='Campaign', aggfunc='count',fill_value=0)
    # reset index
    linelist.reset_index(inplace=True)
    # Joining Form2A Table with Linelist
    linelist = form2a.merge(linelist,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']],right_on='Unique Key')
    # Calculating SMC in Form2A
    linelist['Form2A SMC NA/L'] = (linelist['Recorded Missed NA/L 0-11 M'] + linelist['Recorded Missed NA/L 12-59 M']) - (linelist['Covered Same Day NA/L 0-11 M'] + linelist['Covered Same Day NA/L 12-59 M'])
    linelist['Form2A Refusal'] = (linelist['Recorded Missed Refusal 0-11 M'] + linelist['Recorded Missed Refusal 12-59 M']) - (linelist['Covered Same Day Refusal 0-11 M'] + linelist['Covered Same Day Refusal 12-59 M'])
    #fill nan values with 0
    linelist.fillna(0,inplace=True)
    # Arranging Column
    linelist = linelist.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Form2A SMC NA/L','Form2A Refusal','Not Available','Refused'])
    
    
    # Applying all function here
    linelist.style.\
    apply(na_matching,axis=1,subset=['Form2A SMC NA/L','Not Available']).\
    apply(r_matching,axis=1,subset=['Form2A Refusal','Refused']).to_excel(writer,sheet_name='matching',index=False)
    
elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/zerodose.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    zerodose = pd.read_csv('Data/zerodose.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    
    # ZeroDose
    
    # Filling NaN values with empty
    zerodose.fillna('',inplace=True)
    # Concatenating values
    zerodose['Unique Key'] = zerodose['Campaign'] + zerodose['Province'] + zerodose['District'] + zerodose['Tehsil'] + zerodose['UC'] + zerodose['TID'] + zerodose['Area Incharge'] + zerodose['Day'].astype(str) + zerodose['Team No'].astype(str) + zerodose['Team Member']
    # renaming campaign name to zerodose
    zerodose.rename(columns={'Campaign':'Zerodose'},inplace=True)
    # Transforming data
    zerodose = zerodose.pivot_table(index='Unique Key', values='Zerodose', aggfunc='count',fill_value=0)
    # Reseting index
    zerodose.reset_index(inplace=True)
    # Merging Data with linelist
    zerodose = form2a.merge(zerodose,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    zerodose.fillna(0,inplace=True)
    # Rearraning columns
    zerodose = zerodose.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','Zero Dose Total','Zerodose'])
    #Renaming column name
    zerodose.rename(columns={'Zero Dose Total':'Form2A Zerodose'}, inplace=True)
    
    # Applying all function here
    zerodose.style.apply(zerodose_matching,axis=1,subset=['Form2A Zerodose','Zerodose']).to_excel(writer,sheet_name='matching',index=False)
    
    
elif os.path.exists('Data/form2a.csv') and os.path.exists('Data/afp.csv'):
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    afp = pd.read_csv('Data/afp.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

    
    
    # AFP
    
    # Filling NaN values with empty
    afp.fillna('',inplace=True)
    # Concatenating values
    afp['Unique Key'] = afp['Campaign'] + afp['Province'] + afp['District'] + afp['Tehsil'] + afp['UC'] + afp['TID'] + afp['Area Incharge'] + afp['Day'].astype(str) + afp['Team No'].astype(str) + afp['Team Member']
    # renaming campaign name to zerodose
    afp.rename(columns={'Campaign':'Linelist AFP'},inplace=True)
    # Transforming data
    afp = afp.pivot_table(index='Unique Key', values='Linelist AFP', aggfunc='count',fill_value=0)
    # Reseting index
    afp.reset_index(inplace=True)
    # Merging Data with linelist
    afp = form2a.merge(afp,how='left', left_on=[form2a['Campaign'] + form2a['Province'] + form2a['District'] + form2a['Tehsil'] + form2a['UC'] + form2a['TID'] + form2a['Supervisor Name/Area Incharge'] + form2a['Day'].astype(str) + form2a['CHW Code 1'].astype(str) + form2a['CHW Name 1']], right_on='Unique Key')
    # fill nan values with 0
    afp.fillna(0,inplace=True)
    # Rearraning columns
    afp = afp.reindex(columns=['Campaign','Province','District','Tehsil','UC','TID','Supervisor Name/Area Incharge',
                              'Day','CHW Code 1','CHW Name 1','AFP','Linelist AFP'])
    #Renaming column name
    afp.rename(columns={'AFP':'Form2A AFP', 'Linelist AFP':'AFP'}, inplace=True)
    
    # Applying all function here
    afp.style.apply(afp_matching,axis=1,subset=['Form2A AFP','AFP']).to_excel(writer,sheet_name='matching',index=False)
    

    
elif os.path.exists('Data/form2a.csv'):
    
    # reading file and saving into df varibale
    form2a = pd.read_csv('Data/form2a.csv')
    # fill na with empty
    form2a = form2a.fillna('')
    # filtering Mobil Team Data and repalce it into existing df
    form2a = form2a[form2a['Team Type'] == 'Mobile Team']  
    # validation applying here
    form2a.style.\
        applymap(hh_microplan,subset='Number of HH in Microplan').\
        apply(hh_visited,subset=['Number of HH in Microplan','Number of HH Visited'], axis=1).\
        apply(daily_target_children_1,subset=['Daily Target Children 0-11 M','Vaccinated (A1) 0-11 M','Recorded Missed NA/L 0-11 M','Recorded Missed Refusal 0-11 M'],axis=1).\
        apply(daily_target_children_2,subset=['Daily Target Children 12-59 M','Vaccinated (A1) 12-59 M','Recorded Missed NA/L 12-59 M','Recorded Missed Refusal 12-59 M'],axis=1).\
        applymap(hh_today_1,subset='Children Recorded at HH today 0-11 M').\
        applymap(hh_today_2,subset='Children Recorded at HH today 12-59 M').\
        apply(r1_na,subset=['Recorded Missed NA/L 0-11 M','Covered Same Day NA/L 0-11 M'], axis=1).\
        apply(r2_na,subset=['Recorded Missed NA/L 12-59 M','Covered Same Day NA/L 12-59 M'], axis=1).\
        apply(r1_refusal,subset=['Recorded Missed Refusal 0-11 M','Covered Same Day Refusal 0-11 M'], axis=1).\
        apply(r2_refusal,subset=['Recorded Missed Refusal 12-59 M','Covered Same Day Refusal 12-59 M'], axis=1).\
        apply(c1_na,subset=['Covered Same Day NA/L 0-11 M','Recorded Missed NA/L 0-11 M'], axis=1).\
        apply(c2_na,subset=['Covered Same Day NA/L 12-59 M','Recorded Missed NA/L 12-59 M'], axis=1).\
        apply(c1_refusal,subset=['Covered Same Day Refusal 0-11 M','Recorded Missed Refusal 0-11 M'], axis=1).\
        apply(c2_refusal,subset=['Covered Same Day Refusal 12-59 M','Recorded Missed Refusal 12-59 M'], axis=1).\
        applymap(school_madrassa,subset='School/Madrassa Children').\
        applymap(guest_children,subset='Guest (G) vaccinated').\
        applymap(street_children,subset='Children Vaccinated in Streets').\
        applymap(hrmp,subset='HRMP Children vaccinated (M)').\
        applymap(accute,subset='AFP').\
        applymap(zdose,subset='Zero Dose Total').\
        apply(dose_rec,subset=['Doses Received','Doses Used','Doses Returned'], axis=1).\
        apply(dose_used,subset=['Doses Used','Doses Received'], axis=1).\
        apply(dose_returned,subset=['Doses Returned','Doses Received','Doses Used'], axis=1).\
        apply(dose_vchild,subset=['Vaccinated (A1) 0-11 M','Vaccinated (A1) 12-59 M','Covered Same Day NA/L 0-11 M','Covered Same Day NA/L 12-59 M','Covered Same Day Refusal 0-11 M','Covered Same Day Refusal 12-59 M','School/Madrassa Children','Guest (G) vaccinated','Children Vaccinated in Streets','Doses Used'], axis=1).to_excel(writer,sheet_name='form2a_validation',index=False)

else:
    print('File Not Found\n*Kindly following the instruction to solve this error\n1. Place the files on right path/place\n2. Rename the files as per instructions')
    
```
